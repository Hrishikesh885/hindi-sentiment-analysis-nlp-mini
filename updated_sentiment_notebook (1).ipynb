```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk import ne_chunk
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.svm import SVC
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')

# Load the Hindi SentiWordNet
data = pd.read_csv("HindiSentiWordnet.txt", delimiter=' ')

# Create a dictionary to store the word sentiments
word_sentiments = {}
for i in data.index:
    words = data['LIST_OF_WORDS'][i].split(',')
    for word in words:
        word_sentiments[word] = (data['POS'][i], data['NEG'][i])

# Load the Hindi stopwords
with open("hindi_stopwords.txt", "r") as f:
    hindi_stopwords = f.read().splitlines()

# Load the positive and negative sentences
with open("pos_hindi.txt", "r") as f:
    positive_sentences = f.read().splitlines()

with open("neg_hindi.txt", "r") as f:
    negative_sentences = f.read().splitlines()

# Define a function to preprocess the text
def preprocess_text(text):
    print("Original Text:")
    print(text)

    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    print("Text after removing URLs:")
    print(text)

    # Remove non-Hindi characters
    text = re.sub(r'[^\u0900-\u097F\s]', '', text)
    print("Text after removing non-Hindi characters:")
    print(text)

    # Tokenize the text
    words = word_tokenize(text)
    print("Tokenized Text:")
    print(words)

    # Apply POS tagging
    pos_tags = pos_tag(words)
    print("POS Tags:")
    print(pos_tags)

    # Apply NER
    ner_tags = ne_chunk(pos_tags)
    print("NER Tags:")
    print(ner_tags)

    # Remove stopwords
    words = [word for word, tag in pos_tags if word not in hindi_stopwords]
    print("Text after removing stopwords:")
    print(words)

    # Lemmatize the words
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    print("Text after lemmatization:")
    print(words)

    return ' '.join(words)

# Create a list of sentences and their corresponding labels
sentences = positive_sentences + negative_sentences
labels = [1]*len(positive_sentences) + [0]*len(negative_sentences)

# Preprocess the sentences
preprocessed_sentences = [preprocess_text(sentence) for sentence in sentences]

# Split the data into training and testing sets
train_sentences, test_sentences, train_labels, test_labels = train_test_split(preprocessed_sentences, labels, test_size=0.2, random_state=42)

# Create a CountVectorizer object to convert the text data into numerical features
vectorizer = CountVectorizer()

# Fit the vectorizer to the training data and transform both the training and testing data
train_features = vectorizer.fit_transform(train_sentences)
test_features = vectorizer.transform(test_sentences)

# Create an SVM object and fit it to the training data
svm = SVC()
svm.fit(train_features, train_labels)

# Use the trained SVM to make predictions on the testing data
predictions = svm.predict(test_features)

# Evaluate the accuracy of the SVM
accuracy = svm.score(test_features, test_labels)
print("Accuracy:", accuracy)

# Use the trained SVM to make predictions on new data
def predict_sentiment(sentence):
    preprocessed_sentence = preprocess_text(sentence)
    features = vectorizer.transform([preprocessed_sentence])
    prediction = svm.predict(features)
    return prediction

# Take a Hindi sentence as input from the user
sentence = input("Enter a Hindi sentence: ")

# Use the predict_sentiment function to make a prediction
prediction = predict_sentiment(sentence)
print("Prediction:", prediction)
```

```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk import ne_chunk
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.svm import SVC
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
import ipywidgets as widgets
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')

# Load the Hindi SentiWordNet
data = pd.read_csv("HindiSentiWordnet.txt", delimiter=' ')

# Create a dictionary to store the word sentiments
word_sentiments = {}
for i in data.index:
    words = data['LIST_OF_WORDS'][i].split(',')
    for word in words:
        word_sentiments[word] = (data['POS'][i], data['NEG'][i])

# Load the Hindi stopwords
with open("hindi_stopwords.txt", "r") as f:
    hindi_stopwords = f.read().splitlines()

# Load the positive and negative sentences
with open("pos_hindi.txt", "r") as f:
    positive_sentences = f.read().splitlines()

with open("neg_hindi.txt", "r") as f:
    negative_sentences = f.read().splitlines()

# Define a function to preprocess the text
def preprocess_text(text):
    # Remove URLs
    text = re.sub(r'http\S+', '', text)

    # Remove non-Hindi characters
    text = re.sub(r'[^\u0900-\u097F\s]', '', text)

    # Tokenize the text
    words = word_tokenize(text)

    # Apply POS tagging
    pos_tags = pos_tag(words)

    # Apply NER
    ner_tags = ne_chunk(pos_tags)

    # Remove stopwords
    words = [word for word, tag in pos_tags if word not in hindi_stopwords]

    # Lemmatize the words
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]

    return ' '.join(words)

# Create a list of sentences and their corresponding labels
sentences = positive_sentences + negative_sentences
labels = [1]*len(positive_sentences) + [0]*len(negative_sentences)

# Preprocess the sentences
preprocessed_sentences = [preprocess_text(sentence) for sentence in sentences]

# Split the data into training and testing sets
train_sentences, test_sentences, train_labels, test_labels = train_test_split(preprocessed_sentences, labels, test_size=0.2, random_state=42)

# Create a CountVectorizer object to convert the text data into numerical features
vectorizer = CountVectorizer()

# Fit the vectorizer to the training data and transform both the training and testing data
train_features = vectorizer.fit_transform(train_sentences)
test_features = vectorizer.transform(test_sentences)

# Create an SVM object and fit it to the training data
svm = SVC()
svm.fit(train_features, train_labels)

# Use the trained SVM to make predictions on the testing data
predictions = svm.predict(test_features)

# Evaluate the accuracy of the SVM
accuracy = svm.score(test_features, test_labels)
print("Accuracy:", accuracy)

def analyze_sentiment(text):
    preprocessed_text = preprocess_text(text)
    features = vectorizer.transform([preprocessed_text])
    prediction = svm.predict(features)
    if prediction == 1:
        return "Positive"
    else:
        return "Negative"

text_input = widgets.Text(value='', placeholder='Enter text', description='', disabled=False)
button = widgets.Button(description='Analyze Sentiment')
output = widgets.Text(value='', placeholder='Output', description='', disabled=True)

def on_button_clicked(b):
    text = text_input.value
    output.value = analyze_sentiment(text)

button.on_click(on_button_clicked)

display = widgets.VBox([text_input, button, output])
display
```
